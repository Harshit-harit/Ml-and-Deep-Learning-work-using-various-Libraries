```python
#Extract all the surname from Name column
ship = pd.read_csv('http://bit.ly/kaggletrain')
ship
```

```python
def surname(names):
    for name in names:
        print(name.split(',')[0])

        
surname(ship['Name'])
```

```python
surname =  []
for i in range(len(ship)):
  s = ship['Name'].str.split(',')[i][0]
  surname.append(s)
```

```python
surname =  []
for i in ship['Name']:
  s = i.split(',')[0]
  surname.append(s)
```

```python
ship['surname'] = surname
```
